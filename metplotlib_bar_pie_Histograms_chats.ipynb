# Bar Chart

````python
#plt.bar(x, height, color='color name', width=value, label='label name')

import matplotlib.pyplot as plt

product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

# plt.barh(product, sales, color='orange', label='Sales 2026')  # it gives vertically
plt.bar(product, sales, color='orange', label='Sales 2026')  # it give horizontally
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product Sales Comparision")
plt.legend()
plt.show()
````

# Matplotlib Bar Chart (`plt.bar()`) Quick Notes

## Purpose

A Bar Chart is used to compare values across different categories.

Examples:

* Product Sales
* Student Marks
* Company Revenue
* Population of Cities

---

## Basic Syntax

```python
plt.bar(
    x,
    height,
    color='color_name',
    width=value,
    label='label_name'
)
```

Where:

```python
x       # Categories
height  # Values
```

Example:

```python
product = ['A', 'B', 'C', 'D']
sales = [1000, 1500, 800, 1200]

plt.bar(product, sales)
```

---

## Bar Chart Types

### 1. Vertical Bar Chart

```python
plt.bar(product, sales)
```

Output:

```text
|
|      █
|      █
| █    █
| █    █      █
| █    █  █   █
+----------------
  A    B  C   D
```

Most commonly used.

---

### 2. Horizontal Bar Chart

```python
plt.barh(product, sales)
```

Output:

```text
A ███████
B ███████████
C █████
D ████████
```

Useful when category names are long.

---

### 3. label


Display:

---

### 4. edgecolor

Adds border color.


## Bar vs Barh

### Vertical Bar

```python
plt.bar(product, sales)
```

* Bars grow upward.
* Categories on X-axis.
* Values on Y-axis.

### Horizontal Bar

```python
plt.barh(product, sales)
```

* Bars grow right.
* Categories on Y-axis.
* Values on X-axis.

---

## When to Use Bar Charts

Use Bar Charts when comparing categories:

✅ Product Sales

✅ Student Marks

✅ Company Revenue

✅ Population by City

✅ Monthly Expenses

Avoid Bar Charts for continuous trends over time.

For trends, use:

```python

```

# Matplotlib Pie Chart (`plt.pie()`) Quick Notes

## Purpose

A Pie Chart is used to show the proportion or percentage contribution of each category to the whole.

Examples:

* Revenue by Region
* Market Share
* Budget Distribution
* Population Distribution

---

## Basic Syntax

```python
plt.pie(
    values,
    labels=label_list,
    colors=color_list,
    autopct='%1.1f%%'
)
```

Where:

```python
values   # Numerical values
labels   # Category names
colors   # Slice colors
autopct  # Display percentages
```

Example:

```python
regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions)
```

---

## How Pie Chart Works

Total Revenue:

```python
3000 + 2000 + 1500 + 1000 = 7500
```

Percentage Calculation:

```python
North = (3000/7500)*100 = 40%
South = (2000/7500)*100 = 26.7%
East  = (1500/7500)*100 = 20%
West  = (1000/7500)*100 = 13.3%
```

Pie chart automatically calculates these percentages.

---


# Matplotlib Histogram (`plt.hist()`) Quick Notes

## Purpose

A Histogram is used to show the distribution (frequency) of numerical data.

Examples:

* Student Scores
* Employee Salaries
* Age Distribution
* Height Distribution
* Exam Marks

---

## Basic Syntax

```python
plt.hist(
    data,
    bins=number_of_bins,
    color='color_name',
    edgecolor='black'
)
```

Where:

```python
data       # Numerical data
bins       # Number of intervals/groups
color      # Bar color
edgecolor  # Border color
```

Example:

```python
scores = [45, 67, 89, 56, 78, 88, 92, 60]

plt.hist(scores)
```

---

## What is a Histogram?

A Histogram groups numerical data into ranges and counts how many values fall into each range.

Example:

```python
scores = [23,45,54,56,59,60,65,66,67,74,75,76,78,81,88,89,92,98]
```

Possible Groups:

```text
20-39   → 1 Student
40-59   → 4 Students
60-79   → 8 Students
80-99   → 5 Students
```

Histogram automatically performs this grouping.

---
