# Matplotlib Key Terms You Should Know

## 1. Data Point

A single value plotted on the graph.

Example:

```python
x = [1, 2, 3]
y = [10, 20, 30]
```

Data Points:

```python
(1,10)
(2,20)
(3,30)
```

Each pair `(x,y)` is called a Data Point.

---

## 2. X-axis and Y-axis

### X-axis

* Horizontal axis.
* Usually represents categories, time, months, years, etc.

### Y-axis

* Vertical axis.
* Usually represents values, sales, profit, marks, etc.

Always label both axes.

```python
plt.xlabel("Months")
plt.ylabel("Sales")
```

---

## 3. Figure

The entire canvas/window that contains everything.

Think of it as the whole sheet of paper.

```python
fig = plt.figure()
```

Contains:

* Axes
* Titles
* Legends
* Graphs

---

## 4. Axes

The area where the graph is actually drawn.

Think of it as the graph box inside the figure.

```python
fig, ax = plt.subplots()
```

Figure = Whole Window

Axes = Plotting Area

Visualization:

```text
+------------------------+
|       Figure           |
|                        |
|   +--------------+     |
|   |    Axes      |     |
|   |   Graph      |     |
|   +--------------+     |
|                        |
+------------------------+
```

---

## 5. Plot

The graphical representation of data.

Examples:

* Line Plot
* Bar Plot
* Scatter Plot
* Pie Chart

```python
plt.plot(x, y)
```

---

## 6. Marker

Symbols used to represent data points.

Example:

```python
plt.plot(x, y, marker='o')
```

Common Markers:

```python
'o'   # Circle
's'   # Square
'^'   # Triangle Up
'v'   # Triangle Down
'D'   # Diamond
'*'   # Star
'+'   # Plus
'x'   # Cross
'.'   # Point
```

---

## 7. Line Style

Controls how the line appears.

```python
plt.plot(x, y, linestyle='--')
```

Common Styles:

```python
'-'    # Solid
'--'   # Dashed
'-.'   # Dash Dot
':'    # Dotted
```

---

## 8. Color

Controls the color of lines, markers, and text.

```python
plt.plot(x, y, color='red')
```

Common Colors:

```python
'red'
'blue'
'green'
'black'
'gray'
'yellow'
'purple'
'orange'
```

Short Forms:

```python
'r'
'b'
'g'
'k'
'y'
```

---

## 9. Legend

Explains what each plotted line represents.

Useful when multiple lines exist.

```python
plt.plot(x, y, label="Sales")
plt.legend()
```

Without a legend, users may not know which line represents what.

---

## 10. Label

Descriptive text attached to a graph element.

```python
label='Sales Data'
```

Used inside legends.

---

## 11. Title

The heading of the graph.

```python
plt.title("Monthly Sales Report")
```

Helps users understand what the graph is about.

---

## 12. Grid

Horizontal and vertical reference lines.

```python
plt.grid()
```

Benefits:

* Easier value reading
* Better visualization
* Improved accuracy

---

## 13. Function

A reusable block of code.

Example:

```python
def greet():
    print("Hello")
```

Matplotlib Functions:

```python
plt.plot()
plt.title()
plt.grid()
plt.legend()
```

---

## 14. Method

A function that belongs to an object.

Example:

```python
ax.plot(x, y)
```

Here:

```python
plot()
```

is a method of the Axes object.

---

## 15. Parameter

Inputs accepted by a function.

Example:

```python
plt.plot(x, y, color='red')
```

Parameters:

```python
x
y
color
```

---

## 16. Keyword Arguments

Parameters passed using names.

Example:

```python
plt.plot(
    x,
    y,
    color='red',
    linewidth=2
)
```

Here:

```python
color='red'
linewidth=2
```

are Keyword Arguments.

---

## 17. Object-Oriented API

Professional way of creating plots.

Instead of:

```python
plt.plot(x, y)
```

Use:

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

Advantages:

* Better control
* Cleaner code
* Industry standard
* Used in large projects

---

## 18. DPI (Dots Per Inch)

Controls image quality/resolution.

```python
plt.figure(dpi=100)
```

Examples:

```python
dpi=72    # Low quality
dpi=100   # Normal
dpi=200   # High quality
dpi=300   # Print quality
```

Higher DPI = Better Quality + Larger File Size

---

## 19. Backend

The engine Matplotlib uses to display or save graphs.

Examples:

```python
TkAgg
QtAgg
Agg
inline
```

Common Backends:

### TkAgg

Uses Tkinter GUI.

### QtAgg

Uses Qt GUI.

### Agg

Used for saving images.

### inline

Used in Jupyter Notebook and Google Colab.

Example:

```python
%matplotlib inline
```

Displays plots directly inside notebooks.

---

````python
import matplotlib.pyplot as plt
x = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri'] #x-axis
y = [10,2,30,10,50] #y-axis

plt.plot(x,y)
plt.title("Bakery Sales This Week") # Give heading to the graph
plt.xlabel("Day of the Week")  # x side label
plt.ylabel("Sales per day")    # y side label

plt.show()
````

````python
from matplotlib.lines import lineStyles

# plt.plot(x, y, color='color_name', linestyle='line_style, marker="maker_symbol, label='label_name")

import matplotlib.pyplot as plt

month=[1,2,3,4]
sales=[1000,1200,800,900]

plt.plot(
    month,
    sales,
    color='red',
    linestyle='--',
    linewidth=2,
    marker='o',
    label='2025 Sales Data'
)
plt.title("Monthly Sales Data Report")
plt.xlabel("Months")
plt.ylabel("Sales Per Month")
plt.legend(loc='upper right', fontsize=12) #it tell about the line us cases, it only works when label is given
plt.grid(color='gray', linestyle=':', linewidth=1) # doted lines behind the graph
plt.xlim(1,4) # data limit in x axis
plt.ylim(0,2000) # data limit in y axis
plt.xticks([1,2,3,4], ['M1', 'M2', 'M3', 'M4']) # it replace the original index to given index
plt.yticks()
plt.show()


````

# Matplotlib `plt.grid()` Quick Notes

## Purpose

`plt.grid()` is used to display grid lines on a graph, making it easier to read and compare values.

---

## Basic Syntax

```python
plt.grid()
```

Shows default grid lines on both X and Y axes.

---

## Parameters

### 1. Show / Hide Grid

```python
plt.grid(True)   # Show grid
plt.grid(False)  # Hide grid
```

---

### 2. Grid Color

```python
plt.grid(color='gray')
```

Common colors:

```python
color='red'
color='blue'
color='green'
color='black'
color='gray'
color='yellow'
```

---

### 3. Grid Line Style

```python
plt.grid(linestyle='--')
```

Available styles:

```python
linestyle='-'    # Solid line
linestyle='--'   # Dashed line
linestyle='-.'   # Dash-dot line
linestyle=':'    # Dotted line
```

---

### 4. Grid Line Width

```python
plt.grid(linewidth=2)
```

Examples:

```python
linewidth=0.5    # Thin
linewidth=1      # Normal
linewidth=2      # Thick
linewidth=3      # Very thick
```

---

### 5. Transparency

```python
plt.grid(alpha=0.5)
```

Range:

```python
alpha=0.0    # Invisible
alpha=1.0    # Fully visible
```

---

### 6. Axis Selection

```python
plt.grid(axis='x')     # Only X-axis grid
plt.grid(axis='y')     # Only Y-axis grid
plt.grid(axis='both')  # Both axes
```

---

### 7. Major / Minor Grid

```python
plt.grid(which='major')  # Major grid only
plt.grid(which='minor')  # Minor grid only
plt.grid(which='both')   # Both major and minor
```

---

### 8. Layer Order

```python
plt.grid(zorder=0)
```

* Lower value = Behind graph
* Higher value = Above graph

---

## Complete Example

```python
plt.grid(
    True,
    color='gray',
    linestyle='--',
    linewidth=1,
    alpha=0.7,
    axis='both'
)
```

---

## Most Important Parameters

```python
color
linestyle
linewidth
alpha
axis
```

These are the parameters used most frequently in real-world projects.

---

## Interview Revision

```python

linestyle      # Grid pattern
linewidth      # Grid thickness
alpha          # Transparency
axis           # x, y, both
which          # major, minor, both

```

---

## Commonly Used Grid Command

```python

    color='gray',
    linestyle='--',
    linewidth=1,
    alpha=0.5
)
```

This is the most common grid configuration used in data analysis and machine learning projects.


# Matplotlib `plt.legend()` Quick Notes

## Purpose

`plt.legend()` is used to display labels for plotted lines, bars, or other graph elements.

**Note:** Legend only appears when a `label` is provided.

Example:

```python
plt.plot(x, y, label='Sales')
plt.legend()
```

---

## Basic Syntax

```python
plt.legend()
```

Displays the legend using default settings.

---

## Parameters

### 1. Location (`loc`)

Controls the position of the legend.

```python
plt.legend(loc='upper right')
```

Available values:

```python
loc='best'           # Automatic best location
loc='upper right'    # Top-right
loc='upper left'     # Top-left
loc='lower right'    # Bottom-right
loc='lower left'     # Bottom-left
loc='right'          # Center-right
loc='center left'    # Center-left
loc='center right'   # Center-right
loc='lower center'   # Bottom-center
loc='upper center'   # Top-center
loc='center'         # Center
```

Numeric equivalents:

```python
loc=0    # best
loc=1    # upper right
loc=2    # upper left
loc=3    # lower left
loc=4    # lower right
loc=5    # right
loc=6    # center left
loc=7    # center right
loc=8    # lower center
loc=9    # upper center
loc=10   # center
```

---

### 2. Font Size

Changes legend text size.

```python
plt.legend(fontsize=12)
```

Examples:

```python
fontsize=8
fontsize=10
fontsize=12
fontsize='small'
fontsize='large'
```

---

### 3. Title

Adds a title to the legend.

```python
plt.legend(title='Sales Report')
```

---

### 4. Title Font Size

Changes legend title size.

```python
plt.legend(title_fontsize=14)
```

---

### 5. Frame On/Off

Shows or hides the legend box.

```python
plt.legend(frameon=True)
plt.legend(frameon=False)
```

---

### 6. Background Color

Changes legend background color.

```python
plt.legend(facecolor='lightgray')
```

---

### 7. Border Color

Changes legend border color.

```python
plt.legend(edgecolor='black')
```

---

### 8. Shadow Effect

Adds shadow behind legend.

```python
plt.legend(shadow=True)
```

---

### 9. Number of Columns

Displays legend entries in multiple columns.

```python
plt.legend(ncol=2)
```

---

### 10. Label Spacing

Controls spacing between legend entries.

```python
plt.legend(labelspacing=1.5)
```

---

### 11. Marker Scale

Changes marker size inside legend.

```python
plt.legend(markerscale=2)
```

---

### 12. Border Padding

Controls internal padding of legend box.

```python
plt.legend(borderpad=1)
```

---

### 13. Handle Length

Controls length of line shown in legend.

```python
plt.legend(handlelength=3)
```

---

### 14. Handle Text Padding

Controls spacing between symbol and text.

```python
plt.legend(handletextpad=1)
```

---

### 15. Column Spacing

Controls spacing between columns.

```python
plt.legend(columnspacing=2)
```

---

### 16. Rounded Corners

Creates rounded legend box.

```python
plt.legend(fancybox=True)
```

---

### 17. Transparency

Controls legend box transparency.

```python
plt.legend(framealpha=0.8)
```

Range:

```python
framealpha=0.0   # Invisible
framealpha=1.0   # Fully visible
```

---

### 18. Custom Position

Places legend at a custom location.

```python
plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1)
)
```

Useful for placing legend outside the graph.

---

## Complete Example

```python
plt.legend(
    loc='upper right',
    fontsize=12,
    title='Sales Report',
    title_fontsize=14,
    frameon=True,
    facecolor='lightgray',
    edgecolor='black',
    shadow=True,
    ncol=1
)
```
