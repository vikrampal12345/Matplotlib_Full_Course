# Scattered plot


````python
# plt.scatter(x,y, color='color_name', marker='marker style', label='label name')

import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [67,43,23,90,65,76,89,10]

plt.scatter(hours_studies, exam_scores, color='green', marker='o', label='Student Data')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
# plt.grid(True, alpha=0.3)
plt.show()
````

# Matplotlib Scatter Plot (`plt.scatter()`) Quick Notes

## Purpose

A Scatter Plot is used to show the relationship between two numerical variables.

Examples:

* Study Hours vs Exam Scores
* Height vs Weight
* Age vs Salary
* Advertising Cost vs Sales
* Experience vs Income

---

## Basic Syntax

```python
plt.scatter(
    x,
    y,
    color='color_name',
    marker='marker_style',
    label='label_name'
)
```

Where:

```python
x      # X-axis values
y      # Y-axis values
color  # Point color
marker # Point shape
label  # Legend label
```

Example:

```python
plt.scatter(
    hours_studied,
    exam_scores
)
```

---

## What is a Scatter Plot?

Each point represents one observation.

Example:

```python
hours_studied = [1,2,3]
exam_scores = [50,70,90]
```

Points:

```python
(1,50)
(2,70)
(3,90)
```

Each pair `(x,y)` becomes a dot on the graph.

---

## Important Parameters

### 1. color

Changes point color.

```python
plt.scatter(
    x,
    y,
    color='green'
)
```

Common Colors:

```python
'red'
'blue'
'green'
'orange'
'purple'
'black'
'gray'
'yellow'
```

---

### 2. marker

Changes point shape.

```python
plt.scatter(
    x,
    y,
    marker='o'
)
```

Common Markers:

```python
'o'   # Circle
's'   # Square
'^'   # Triangle Up
'v'   # Triangle Down
'D'   # Diamond
'*'   # Star
'+'   # Plus
'x'   # Cross
'.'   # Point
```

---

### 3. label

Used for legend.

```python
plt.scatter(
    x,
    y,
    label='Student Data'
)

plt.legend()
```

---

### 4. s (Size)

Controls marker size.

```python
plt.scatter(
    x,
    y,
    s=100
)
```

Examples:

```python
s=20
s=50
s=100
s=200
```

---

### 5. alpha

Controls transparency.

```python
plt.scatter(
    x,
    y,
    alpha=0.5
)
```

Range:

```python
alpha=0.0   # Invisible
alpha=1.0   # Fully Visible
```

---

### 6. edgecolors

Adds border color to markers.

```python
plt.scatter(
    x,
    y,
    edgecolors='black'
)
```

---

### 7. linewidths

Controls marker border thickness.

```python
plt.scatter(
    x,
    y,
    linewidths=2
)
```

---

### 8. cmap

Applies color map when using multiple colors.

```python
plt.scatter(
    x,
    y,
    c=y,
    cmap='viridis'
)
```

Common Colormaps:

```python
viridis
plasma
cool
hot
rainbow
```

---

### 9. c

Assigns colors using values.

```python
plt.scatter(
    x,
    y,
    c=y
)
```

Useful for advanced visualizations.

---

## Complete Example

```python
import matplotlib.pyplot as plt

hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [67,43,23,90,65,76,89,10]

plt.scatter(
    hours_studied,
    exam_scores,
    color='green',
    marker='o',
    s=100,
    label='Student Data'
)

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```

## Understanding Correlation

### Positive Correlation

```text

  •
    •
      •
```

As X increases, Y increases.

Example:

```python
Study Hours ↑

```

---

### Negative Correlation

```text
      •
    •
  •
•
```

As X increases, Y decreases.

Example:

```python
Speed ↑

```

---

### No Correlation

```text
•      •
   •
      •
 •
```

No clear pattern.

## When to Use Scatter Plot

Use Scatter Plot for:

✅ Machine Learning

✅ Data Exploration

---

## When NOT to Use Scatter Plot

Avoid Scatter Plot for:

❌ Category Comparison

Use:

```python
plt.bar()
```

❌ Percentage Distribution

Use:

```python
plt.pie()
```

❌ Frequency Distribution

Use:

```python
plt.hist()
```

## Most Common Scatter Plot

```python
plt.scatter(
    x,
    y,
    color='green',
    marker='o',
    label='Data'
)
```

This is the most commonly used scatter plot configuration in Data Analysis, Statistics, and Machine Learning projects.


````python
import matplotlib.pyplot as plt

hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [67,43,23,90,65,76,89,99]
hours_studies_1 = [1,2,3,4,5,6,7,8]
exam_scores_2= [60,50,20,92,60,70,95,100]

plt.scatter(hours_studies, exam_scores, color='green', marker='o', label='Student Data 1')
plt.scatter(hours_studies_1, exam_scores_2, color='red', marker='o', label='Student Data 2')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Comparision between two groups')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

# Multiple Scatter Plot Comparison Notes

## Purpose

Multiple Scatter Plots are used to compare two or more datasets on the same graph.

Example:

```python
plt.scatter(x1, y1, label='Group 1')
plt.scatter(x2, y2, label='Group 2')
```

---

## New Keywords Introduced

### Multiple `plt.scatter()`

```python
plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

Used to plot multiple datasets on the same graph.

---

### Different Colors

```python
color='green'
color='red'
```

Helps distinguish different groups.

---

### Different Labels

```python
label='Student Data 1'
label='Student Data 2'
```

Used by legend to identify groups.

---

### Legend Comparison

```python
plt.legend()
```

Shows:

```text
Green → Student Data 1
Red   → Student Data 2
```

---

## Comparison Use Cases

### Student Group Comparison

```text
Group A vs Group B
```

---

### Product Comparison

```text
Product A vs Product B
```

---

### Before vs After Analysis

```text
Before Training
After Training
```

---

### Machine Learning Classes

```text
Class 0
Class 1
```

---

### Cluster Comparison

```text
Cluster A
Cluster B
Cluster C
```

---

## Important Observation

If points overlap:

```python
alpha=0.5
```

Makes overlapping points visible.

---

If groups are hard to identify:

```python
marker='o'
marker='s'
```

Use different marker styles.

Example:

```python
plt.scatter(x1, y1, marker='o')
plt.scatter(x2, y2, marker='s')
```

---

## Most Common Comparison Pattern

```python
plt.scatter(
    x1,
    y1,
    color='green',
    label='Group 1'
)

plt.scatter(
    x2,
    y2,
    color='red',
    label='Group 2'
)

plt.legend()
```
